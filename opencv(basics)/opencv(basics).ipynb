```python
import cv2
```

```python
img = cv2.imread ('/content/download.jpeg',1)   # Here, path can be either the path of image in your device, in Colab or a image URL
 
# For Black and White (gray scale)
# img = cv2.imread (path,0)
```

```python
img.shape
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)

# waiting for key event
cv2.waitKey(0)

# destroying all windows
cv2.destroyAllWindows()
```

```python
# Resizing the image


res_img = cv2.resize(img,(400,400))

cv2_imshow(res_img)
```

```python
# -------   GREY SCALE    ----
# Converting color image to grayscale image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Showing the converted image

# Colab users:
cv2_imshow(gray_img)


# waiting for key event
cv2.waitKey(0)

# destroying all windows
cv2.destroyAllWindows()
```
